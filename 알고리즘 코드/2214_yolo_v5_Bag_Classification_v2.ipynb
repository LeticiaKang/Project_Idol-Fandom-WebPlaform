```python
# Dataset ë‚´ ì´ë¯¸ì§€ ê°œìˆ˜ í™•ì¸

from glob import glob

img_list = glob('/content/gdrive/MyDrive/Colab Notebooks/yolo_v5_Bag_Classification_v2/Dataset_bag/images/*')
print(len(img_list))
```

```python
# sklearn => train, test data ë‚˜ëˆ„ê¸°

from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size = 0.3, random_state = 2000, shuffle = True) # 20%ì˜ ë°ì´í„°ë¥¼ validation dataë¡œ
                                                                                                # random_stateëŠ” ì„ì˜ë¡œ ì§€ì •

print(len(train_img_list), len(val_img_list))
```

```python
# Training image, validation image ê²½ë¡œë¥¼ .txt íŒŒì¼ë¡œ ì €ì¥
with open('/content/gdrive/MyDrive/Colab Notebooks/yolo_v5_Bag_Classification_v2/Dataset_bag/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/gdrive/MyDrive/Colab Notebooks/yolo_v5_Bag_Classification_v2/Dataset_bag/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
# ì´ì œ Custom_data.yaml íŒŒì¼ì„ ë§Œì ¸ì¤˜ì•¼ í•¨ 

import yaml

with open('/content/gdrive/MyDrive/Colab Notebooks/yolo_v5_Bag_Classification_v2/Dataset_bag/Custom_data.yaml', 'r') as f:
  data = yaml.load(f,Loader= yaml.FullLoader) # Custom_data.yaml íŒŒì¼ ì½ê¸°

print(data)

data['train'] = '/content/gdrive/MyDrive/Colab Notebooks/yolo_v5_Bag_Classification_v2/Dataset_bag/train.txt' # train dataì˜ ê²½ë¡œ ì§€ì • ë° train.txtë¡œ ìƒì„±
data['val'] = '/content/gdrive/MyDrive/Colab Notebooks/yolo_v5_Bag_Classification_v2/Dataset_bag/val.txt'# val dataì˜ ê²½ë¡œ ì§€ì • ë° train.txtë¡œ ìƒì„±

with open('/content/gdrive/MyDrive/Colab Notebooks/yolo_v5_Bag_Classification_v2/Dataset_bag/Custom_data.yaml', 'w') as f: #custom
  yaml.dump(data, f)

print(data)
```
